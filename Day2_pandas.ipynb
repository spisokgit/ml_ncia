
## 정규화(minmax), 표준화(stand):표준정규분포

```python
#교재 162 참조
#최소_최대 정규화 # sklearn MinMaxScaler  ex) .fit
def minmax_normalizer(arr):
    return (arr-arr.min())/(arr.max()-arr.min())

# z-점수 표준화   # StandScaler
def zscore_standize(arr):
    return(arr - arr.mean())/arr.std()
```

```python
cnum = concrete.iloc[:8, :-1][:5]
```

```python
cnum.plot()
```

```python
cnum.apply(minmax_normalizer).plot() # 
```

```python
cnum.apply(zscore_standize).plot()
```
